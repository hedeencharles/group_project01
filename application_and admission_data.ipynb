```python
# import dependancies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import time
import json
```

```python
# import college applications and admissions data from IPEDS
application_df = pd.read_csv("college_app_data.csv")
application_df.head()
```

```python
# clean up dataframe to remove 2001 and 2019
clean_app_df = application_df.loc[(application_df['Year'] >=2002) & 
                                  (application_df['Year'] <=2018)
                                 ]
clean_app_df.head()
```

```python
# sort by college names and year
application_df_sorted= clean_app_df.sort_values(["College_names","Year"],ascending = (True))
# application_df_sorted.head()
# calculate applicants change YOY and admissions change YOY
application_df_sorted['% applicants change'] =application_df_sorted.groupby('College_names').applicants_total.pct_change()
application_df_sorted['% admissions_change'] =application_df_sorted.groupby('College_names').admissions_total.pct_change()
application_df_sorted.head(5)
```

```python
# calculate difference between applicants and admissions
application_df_sorted['applications_vs_admissions'] = application_df_sorted['admissions_total'] - application_df_sorted['applicants_total']
application_df_sorted.head()
```

```python
# export clean application and admissions data
application_df_sorted.to_csv("clean_app_data.csv", index=False, header = True)
```
