```python
application_df = pd.read_csv("college_app_data.csv")
application_df.head()

# application_df_sorted = application_df.sort_values(["College_names","Year"],ascending = (True))
```

```python
application_df_sorted= application_df.sort_values(["College_names","Year"],ascending = (True))
# application_df_sorted.head()
application_df_sorted['% applicants change'] =application_df_sorted.groupby('College_names').applicants_total.pct_change()
application_df_sorted['% admissions_change'] =application_df_sorted.groupby('College_names').admissions_total.pct_change()
application_df_sorted.head(20)
```
